```python
df['Gender'] = 4
df.head()
```

```python
df['Gender'] = df.Sex.map(lambda x: x[0].upper())
df.head()
```

```python
df['Gender'] = df.Sex.map({'female': 0, 'male': 1}).astype(int)
df['Embark'] = df.Embarked.map({'S': 0, 'C':1, 'Q':2, None: -1})
df[df.Embarked.isnull()]
```

```python
median_ages = np.zeros((2, 3))
median_ages
```

```python
for i in range(2):
    for j in range(3):
        median_ages[i,j] = df[(df.Gender == i) & (df.Pclass == j+1)].Age.dropna().median()
median_ages
```

```python
df['AgeFill'] = df.Age
df[['Pclass', 'Gender', 'Age', 'AgeFill']][df.Age.isnull()].head(15)
```

```python
for i in range(2):
    for j in range(3):
        df.loc[(df.Gender == i) & (df.Pclass == j+1) & (df.Age.isnull()), 'AgeFill'] = median_ages[i,j]
df[(df.Age.isnull())].head()
```

```python
df['AgeIsNull'] = df.Age.isnull().astype(int)
df.head()
```

```python
df.describe()
```

```python
df['FamilySize'] = df['SibSp'] + df['Parch']
df['Age*Class'] = df.AgeFill * df.Pclass
df.FamilySize.hist()
```

```python
df = df.drop(['Name', 'Sex', 'Ticket', 'Cabin', 'Embarked', 'Age'], axis=1)
```

```python
df = df.dropna()
df.head()
```
